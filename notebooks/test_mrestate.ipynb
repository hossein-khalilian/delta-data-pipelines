```python
import re
import httpx
from curl2json import parser
from bs4 import BeautifulSoup
```

```python
curl_command = """curl 'https://mrestate.ir/' \
  -I \
  -H 'User-Agent: Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:145.0) Gecko/20100101 Firefox/145.0' \
  -H 'Accept: */*' \
  -H 'Accept-Language: en-US,en;q=0.5' \
  -H 'Accept-Encoding: gzip, deflate, br, zstd' \
  -H 'Referer: https://mrestate.ir/f/all' \
  -H 'x-middleware-preflight: 1' \
  -H 'Sec-GPC: 1' \
  -H 'Connection: keep-alive' \
  -H 'Cookie: key_search_history=3605235686' \
  -H 'Sec-Fetch-Dest: empty' \
  -H 'Sec-Fetch-Mode: cors' \
  -H 'Sec-Fetch-Site: same-origin' \
  -H 'Priority: u=4'"""
```

```python
parsed_curl = parser.parse_curl(curl_command)
```

```python
resp = httpx.request(**parsed_curl)
html = resp.content
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
import json
next_data_script = soup.find("script", id="__NEXT_DATA__")

if next_data_script:
    data = json.loads(next_data_script.string)
    build_id = data.get("buildId")
    print("Build ID:", build_id)
else:
    print("__NEXT_DATA__ tag not found.")
```
